```python
import pandas as pd

# Load the timetable workbook (relative path — works on any machine after cloning the repo)
excel_file_path = "timetable-fsc-fall-2023.xlsx"
xls = pd.ExcelFile(excel_file_path, engine="openpyxl")
print("Available sheets:", xls.sheet_names)
```

```python
# Read the Monday sheet with no header assumption first, since the sheet
# has a multi-row header block (title + degree-program legend + time-slot row)
df_raw = pd.read_excel(xls, "Monday", header=None)

# Row 4 (0-indexed) is the actual header row: "Room" + one label per time slot
header_row_idx = 4
room_col = 0
time_row = df_raw.iloc[header_row_idx]

# Each time slot spans 3 merged columns starting at column 1
time_slots = []
col = 1
while col < df_raw.shape[1]:
    label = time_row[col]
    if pd.notna(label):
        time_slots.append((col, str(label).strip()))
    col += 1

time_slots
```

```python
# Build a long-format dataframe: index, day, room, time, course, type
records = []
idx = 0
for r in range(header_row_idx + 1, df_raw.shape[0]):
    room = df_raw.iat[r, room_col]
    if pd.isna(room):
        continue
    for start_col, time_label in time_slots:
        # Each time slot spans up to 3 columns; take the first non-empty cell
        span = df_raw.iloc[r, start_col:start_col + 3]
        val = next((v for v in span if pd.notna(v)), None)
        if val:
            course = str(val).strip()
            course_type = "Lab" if "lab" in course.lower() else "Theory"
            records.append({
                "index": idx,
                "day": "Monday",
                "room": str(room).strip(),
                "time": time_label,
                "course": course,
                "type": course_type
            })
            idx += 1

monday_df = pd.DataFrame(records)
monday_df.head(15)
```
